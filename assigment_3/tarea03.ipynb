1. Implementar backtracking para $\alpha$ inexacto, condición de armijo

```python
def back_tracking(alpha_init:float, 
                  rho:float,
                  c:float, 
                  xk:np.ndarray,
                  f:Callable[[np.ndarray],float],
                  fk:float,
                  grad_fk:np.ndarray,
                  dir_pk:np.ndarray,
                  iter_maxb:int=100)->Tuple[float,int]:
    
    """Backtracking line search algorithm using Armijo condition
    
    :param alpha_init:  initial step length (float)
    :param rho: decay factor for step length (float)
    :param c: constant for Armijo condition (float)
    :param xk: current point (np.ndarray)
    :param f: objective function (Callable[[np.ndarray],float])
    :param fk: objective function value at xk (float)
    :param grad_fk: gradient of objective function at xk (np.ndarray)
    :param dir_pk: search direction (np.ndarray)
    :param iter_maxb: maximum number of iterations for backtracking (int)
    
    :return: step length and number of iterations (Tuple[float,int]) 
    """
    
    alpha = alpha_init
    k = 0
    while k < iter_maxb:
        if f(xk + alpha*dir_pk) <= fk + c*alpha*np.dot(grad_fk,dir_pk):
            return alpha, k
        alpha *= rho
        k += 1
    print('Backtracking line search did not converge')
    return alpha, k
```

2. Implementar descenso máximo con Backtracking

```python
def gradient_descent_backtracking(  f:Callable[[np.ndarray],float],
                                    grad_f:Callable[[np.ndarray],np.ndarray],
                                    x0:np.ndarray,
                                    tol:float,
                                    iter_max:int,
                                    alpha_init:float,
                                    rho:float,
                                    c:float,
                                    max_iter_bt:int=100)->Tuple[np.ndarray,int,bool]:
    
    """Gradient descent with backtracking line search
    
    :param f: objective function (Callable[[np.ndarray],float])
    :param grad_f: gradient of objective function (Callable[[np.ndarray],np.ndarray])
    :param x0: initial point (np.ndarray)
    :param tol: tolerance for stopping criterion (float)
    :param iter_max: maximum number of iterations (int)
    :param alpha_init: initial step length (float)
    :param rho: decay factor for step length (float)
    :param c: constant for Armijo condition (float)
    :param iter_max_bt: maximum number of iterations for backtracking line search (int)
    
    :return: optimal point, number of iterations, True if gradient descent converges (Tuple[np.ndarray,int,bool])
    """

    x = x0
    k = 0
    n = x0.shape[0]
    list_xk = np.array([])
    while k < iter_max:
        grad_fk = grad_f(x)
        fk = f(x)
        dir_pk = -grad_fk
        if n==2:
            alpha, k_bt = back_tracking(alpha_init, rho, c, x, f, fk, grad_fk, dir_pk, max_iter_bt)
            list_xk = np.append(list_xk, [x[0], x[1], alpha, k_bt])
        else:
            alpha, _ = back_tracking(alpha_init, rho, c, x, f, fk, grad_fk, dir_pk, max_iter_bt)
        
        if alpha*np.linalg.norm(grad_fk) < tol:
            if n==2:
                return x, k, True, list_xk.reshape(-1,4)
            else:
                return x, k, True
        x = x + alpha*dir_pk
        k += 1
    if n==2:
        return x,k,False,list_xk.reshape(-1,4)
    print('Gradient descent did not converge')
    return x, k,False
```

Implementación del visualizador de las trayectorias obtenidas en el algoritmo descenso máximo por backtracking

```python
def contornosFnc2D(fncf, xleft, xright, ybottom, ytop, levels, list_xk1, list_xk2=None):
    # Crea una discretización uniforme del intervalo [xleft, xright]
    ax = np.linspace(xleft, xright, 250)
    # Crea una discretización uniforme del intervalo [ybottom, ytop]
    ay = np.linspace(ybottom, ytop, 200)
    # La matriz mX que tiene las abscisas 
    mX, mY = np.meshgrid(ax, ay)
    # Se crea el arreglo mZ con los valores de la función en cada nodo
    mZ = mX.copy()
    for i,y in enumerate(ay):
        for j,x in enumerate(ax):
            mZ[i,j] = fncf(np.array([x,y]))
    # Grafica de las curvas de nivel
    fig, ax = plt.subplots(figsize=(8,8))
    CS = ax.contour(mX, mY, mZ, levels, cmap='Wistia')
    
    ax.plot(list_xk1[:,0], list_xk1[:,1], marker='o', linestyle='-', color='blue', label='Trajectory $Xk_1$: x0={}'.format(list_xk1[0]), markersize=4)
    if list_xk2 is not None:
        ax.plot(list_xk2[:,0], list_xk2[:,1], marker='o', linestyle='-', color='red', label='Trajectory $Xk_2$: x0={}'.format(list_xk2[0]), markersize=4)
    ax.legend()
```

**Función de Himmelblau:** 

Para $\mathbf{x}=(x_1,x_2)$, 

$$\nabla f(\mathbf{x}) = (x_1^2 + x_2 - 11)^2 + (x_1 + x_2^2 - 7)^2. $$

donde su gradiente es:


$$\Delta f(\mathbf{x}) = (4x_1(x_1^2 + x_2 - 11)+2(x_1 + x_2^2 - 7),2(x_1^2 + x_2 - 11)+4x_2(x_1 + x_2^2 - 7)) $$

```python
def fncHimmelblau(x):
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2

def grad_fncHimmelblau(x):
    return np.array([4*x[0]*(x[0]**2 + x[1] - 11) + 2*(x[0] + x[1]**2 - 7), 
                     2*(x[0]**2 + x[1] - 11) + 4*x[1]*(x[0] + x[1]**2 - 7)])
```

```python
x01 = np.zeros(2)
tol = np.sqrt(x01.shape[0]*np.finfo(float).eps)
xk1, k1, conv1, list_xk1 = gradient_descent_backtracking(fncHimmelblau,grad_fncHimmelblau,x01,tol,max_iter,alpha_init,rho,c,max_iter_bt)

print("Himmelblau function\n")

print("Valor inicial x0\t:",x01)
print("Número de iteraciones\t:",k1)
print("xk\t\t\t:",xk1)
print("f(xk)\t\t\t:",fncHimmelblau(xk1))
print("||grad_f(x_k)|| \t:",np.linalg.norm(grad_fncHimmelblau(xk1)))
print("Convergencia\t\t:",conv1)
print("Prom. tamaño de paso\t:",round(np.mean(list_xk1[:,2]),4))
print("Prom. iteraciones de bt\t:",round(np.mean(list_xk1[:,3]),4), "\n")

x02 = np.array([2,4])
tol = np.sqrt(x02.shape[0]*np.finfo(float).eps)
xk2, k2, conv2, list_xk2 = gradient_descent_backtracking(fncHimmelblau,grad_fncHimmelblau,x02,tol,max_iter,alpha_init,rho,c,max_iter_bt)

print("Valor inicial x0\t:",x02)
print("Número de iteraciones\t:",k2)
print("xk\t\t\t:",xk2)
print("f(xk)\t\t\t:",fncHimmelblau(xk2))
print("||grad_f(x_k)|| \t:",np.linalg.norm(grad_fncHimmelblau(xk2)))
print("Convergencia\t\t:",conv2)
print("Prom. tamaño de paso\t:",round(np.mean(list_xk2[:,2]),4))
print("Prom. iteraciones de bt\t:",round(np.mean(list_xk2[:,3]),4), "\n")

contornosFnc2D(fncHimmelblau, xleft=-0.5, xright=5.5, ybottom=-0.5, ytop=5.5,
               levels=[0.5,2, 5, 10, 25, 50, 100, 150, 250, 400],list_xk1=list_xk1[:,:2], list_xk2=list_xk2[:,:2])
```

**Función de Beale :** Para $\mathbf{x}=(x_1,x_2)$

```python
def fncBeale(x):
    return (1.5-x[0]+x[0]*x[1])**2+(2.25-x[0]+x[0]*x[1]**2)**2+(2.625-x[0]+x[0]*x[1]**3)**2
def grad_fncBeale(x):
    return np.array([2*(1.5-x[0]+x[0]*x[1])*(-1+x[1])+2*(2.25-x[0]+x[0]*x[1]**2)*(-1+x[1]**2)+2*(2.625-x[0]+x[0]*x[1]**3)*(-1+x[1]**3),
                     2*(1.5-x[0]+x[0]*x[1])*x[0]+4*(2.25-x[0]+x[0]*x[1]**2)*x[0]*x[1]+6*(2.625-x[0]+x[0]*x[1]**3)*x[0]*x[1]**2])
```

**Función de Rosenbrock:** Para $\mathbf{x}=(x_1,x_2, ..., x_n)$

$$ f(\mathbf{x}) = \sum_{i=1}^{n-1} \left[100(x_{i+1} - x_i^2)^2 + (1-x_i)^2 \right]
\quad n\geq 2.$$


```python
def fncRosenbrock(x):
    return np.sum(100*(x[1:]-x[:-1]**2)**2+(1-x[:-1])**2)

def grad_fncRosenbrock(x):
    grad = np.zeros(len(x))
    grad[0] = -400*x[0]*(x[1]-x[0]**2)+2*x[0]-2
    grad[-1] = 200*(x[-1]-x[-2]**2)
    for i in range(1,len(x)-1):
        grad[i] = 200*(x[i]-x[i-1]**2)-400*x[i]*(x[i+1]-x[i]**2)+2*x[i]-2
    return grad
```

La desigualdad de armijo nos indica que la función debe descender en la siguiente iteración proporcionalmente al tamaño del paso y la derivada direccional de la dirección de descenso

$$ f(\mathbf{x}_0 + \alpha\mathbf{p}_0) \leq 
f(\mathbf{x}_0) + c_1 \alpha\mathbf{p}_0^\top\nabla f(\mathbf{x}_0). $$ 

La condición de Armijo, nos quiere decir que $c_1$ controla que tanto debe disminuir la función en la siguiente iteración para considerarse que descendio. Conviene valores grandes de $c_1$, es decir, valores cercanos a 1, cuando tengamos conocimiento de que la función disminuye bastante en las direcciones de descenso, ya que si se acerca a uno los puntos que satisfacen la condición la función debe disminuir casi o más que lo que lo hace la aproximación lineal, lo cual no se aconseja en vecindades muy proximas a la función si tiene un comportamiento convexo, el otro caso, nos da mayor libertad de que la función no debe disminuir mucho, valores de $c_1$ pequeños se recomiendan cuando se está cerca al mínimo, ya que la función no va a presentar cambios tan marcados.

**Solución:**

```python
def gradient_descent_backtracking_2(f:Callable[[np.ndarray],float], 
                                    grad_f:Callable[[np.ndarray],np.ndarray],
                                    x0:np.ndarray, 
                                    tol:float, 
                                    iter_max:int, 
                                    alpha_init:float, 
                                    rho:float, 
                                    c:float,max_iter_bt)->Tuple[np.ndarray,int,bool,np.ndarray]:
    """Gradient descent with backtracking line search modified 
    
    :param f: objective function (Callable[[np.ndarray],float])
    :param grad_f: gradient of objective function (Callable[[np.ndarray],np.ndarray])
    :param x0: initial point (np.ndarray)
    :param tol: tolerance for stopping criterion (float)
    :param iter_max: maximum number of iterations (int)
    :param alpha_init: initial step length (float)
    :param rho: decay factor for step length (float)
    :param c: constant for Armijo condition (float)
    :param max_iter_bt: maximum number of iterations for backtracking line search (int)
    
    :return: optimal point, number of iterations, True if gradient descent converges, list of points (Tuple[np.ndarray,int,bool,np.ndarray])
    """
    
    x = x0
    k = 0
    n = x0.shape[0]
    list_xk = np.array([])
    while k < iter_max:
        if k>0:
            s = np.dot(grad_fk,dir_pk)
        grad_fk = grad_f(x)
        fk = f(x)
        dir_pk = -grad_fk
        if n==2:
            if k==0:
                alpha, k_bt = back_tracking(alpha_init, rho, c, x, f, fk, grad_fk, dir_pk, max_iter_bt)
                list_xk = np.append(list_xk, [x[0], x[1], alpha, k_bt])
            else:
                alpha_new = alpha*min(100,s/np.dot(grad_fk,dir_pk))
                alpha, k_bt = back_tracking(alpha_new, rho, c, x, f, fk, grad_fk, dir_pk, max_iter_bt)
                list_xk = np.append(list_xk, [x[0], x[1], alpha, k_bt])
        else:
            if k==0:
                alpha, _ = back_tracking(alpha_init, rho, c, x, f, fk, grad_fk, dir_pk, max_iter_bt)
            else:
                alpha_new = alpha*min(100,s/np.dot(grad_fk,dir_pk))
                alpha, _ = back_tracking(alpha_new, rho, c, x, f, fk, grad_fk, dir_pk, max_iter_bt)
            
        if alpha*np.linalg.norm(grad_fk) < tol:
            if n==2:
                return x, k, True, list_xk.reshape(-1,4)
            else:
                return x, k, True
        x = x + alpha*dir_pk
        k += 1
    
    if n==2:
        return x,k,False,list_xk.reshape(-1,4)
    print('Gradient descent did not converge')
    return x, k,False
```

# <font color="8C3061" >**Ejercicio 4 (1.5 puntos)**</font>

Sea $f:\mathbb{R}^n \rightarrow \mathbb{R}$ y $\mathbf{S}$ una matriz no singular de tamaño $n \times n$. 
Si $\mathbf{x} = \mathbf{S}\mathbf{y}$ para $\mathbf{y} \in \mathbb{R}^n$ y definimos
$g(\mathbf{y}) = f(\mathbf{S}\mathbf{y})$, aplicando la regla de la cadena muestre que 

$$ \nabla g(\mathbf{y}) =  \mathbf{S}^\top \nabla f(\mathbf{x}). $$

Entonces aplicando el método de máximo descenso a la función $g$ es 

$$ \mathbf{y}_{k+1} = \mathbf{y}_k - \alpha_k \mathbf{S}^\top \nabla f(\mathbf{S}\mathbf{y}_k).$$

Multiplicando por $\mathbf{S}$ ambos miembros de la ecuación y usando la notación
$\mathbf{x}_k =\mathbf{S} \mathbf{y}_k$:

$$ \mathbf{x}_{k+1} =  \mathbf{S} \mathbf{y}_{k+1} = 
\mathbf{S} \mathbf{y}_k - \alpha_k \mathbf{S}\mathbf{S}^\top \nabla f(\mathbf{S}\mathbf{y}_k)
= \mathbf{x}_k - \alpha_k \mathbf{S}\mathbf{S}^\top \nabla f(\mathbf{x}_k).$$

Si $\mathbf{D} = \mathbf{S}\mathbf{S}^\top$, obtenemos el método de *gradiente escalado*:

$$ \mathbf{x}_{k+1} = \mathbf{x}_k - \alpha_k \mathbf{D} \nabla f(\mathbf{x}_k).$$

Muestre que $-\mathbf{D} \nabla f(\mathbf{x}_k)$ es una dirección de descenso.

**Solución:**

Tenemos que podemos expresar la función $g:\mathbb{R}^n\rightarrow \mathbb{R}$, como $g(\mathbf{y})=f(S(\mathbf{y}))$, así por la regla de la cadena obtenemos que

$$\nabla g(\mathbf{y})=\nabla f(\mathbf{S}(\mathbf{y})) \nabla(\mathbf{S}(\mathbf{y}))=\nabla f(\mathbf{x})^\top\mathbf{S}$$

Lo que es equivalente a

$$\nabla g(\mathbf{y}) = \mathbf{S}^\top \nabla f(\mathbf{x})$$

Ahora bien, para ver que $-\mathbf{D} \nabla f(\mathbf{x}_k)$ es una dirección de descenso, es suficiente ver que su producto punto con el grafiente de $f$ es negativo

$$
\begin{align*}
\langle-\mathbf{D} \nabla f(\mathbf{x}_k),\nabla f(\mathbf{x}_k)\rangle&=-\langle\mathbf{D} \nabla f(\mathbf{x}_k),\nabla f(\mathbf{x}_k)\rangle\\
&=-\nabla f(\mathbf{x}_k)^\top\mathbf{D} \nabla f(\mathbf{x}_k)\\
&=-\nabla f(\mathbf{x}_k)^\top\mathbf{S}\mathbf{S}^\top \nabla f(\mathbf{x}_k)\\
&=-(\mathbf{S}^\top\nabla f(\mathbf{x}_k))^\top(\mathbf{S}^\top\nabla f(\mathbf{x}_k))\\
&=-\|(\mathbf{S}^\top\nabla f(\mathbf{x}_k))\|_2^2\\
&\leq 0
\end{align*}
$$

Por tanto, podemos concluír que $-\mathbf{D} \nabla f(\mathbf{x}_k)$ es una dirección de descenso.
